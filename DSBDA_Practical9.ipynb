```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
#titanic dataset
data = pd.read_csv("titanic_train.csv")
#tips dataset
tips = load_dataset("tips")
```

```python
sns.countplot(x='Survived', data=data)
plt.show()
```

```python
data['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(data['Age'], bins=5)
plt.show()
```

```python

sns.displot(data['Age'], kde=True)
plt.show()
```

```python
sns.boxplot(x=data['Age'])

plt.title("Boxplot of Age")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Scatter Plot of Total Bill vs. Tip")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)

plt.title("Scatter Plot of Total Bill vs. Tip by Gender")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)

plt.title("Scatter Plot of Total Bill vs. Tip by Gender and Smoker Status")
plt.show()
```

```python
sns.barplot(x="Pclass", y="Age", data=data)

plt.title("Average Age by Passenger Class")
plt.show()
```

```python
sns.barplot(x="Pclass", y="Fare", hue="Sex", data=data)

plt.title("Average Fare by Passenger Class and Gender")
plt.show()
```

```python
sns.boxplot(x="Sex", y="Age", data=data)

plt.title("Boxplot of Age by Gender")
plt.show()
```

```python
sns.boxplot(x="Sex", y="Age", hue="Survived", data=data)

plt.title("Boxplot of Age by Gender and Survival Status")
plt.show()
```

```python
sns.kdeplot(data=data[data['Survived'] == 0]['Age'], color="blue", label="Non-Survivors", fill=True)
sns.kdeplot(data=data[data['Survived'] == 1]['Age'], color="red", label="Survivors", fill=True)
plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
pd.crosstab(data['Pclass'], data['Survived'])
```

```python
sns.heatmap(pd.crosstab(data['Pclass'], data['Survived']))
```

```python
sns.clustermap(pd.crosstab(data['Parch'], data['Survived']))
plt.show()
```
