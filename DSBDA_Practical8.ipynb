```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = sns.load_dataset('iris')
dataset.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
plt.figure(figsize=(8, 5))
sns.histplot(dataset['sepal_length'], kde=True, bins=20, color="blue")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.displot(dataset['sepal_length'], kde=True, color='blue')
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.jointplot(x='sepal_length', y='sepal_width', data=dataset, kind='scatter')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.rugplot(dataset['sepal_length'], color='blue')
plt.title("Rug Plot of Sepal Length")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.barplot(x='species', y='sepal_length', data=dataset)
plt.title("Average Sepal Length by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.countplot(x='species', data=dataset)
plt.title("Count of Each Species in Iris Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.boxplot(x='species', y='sepal_length', data=dataset)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.violinplot(x='species', y='sepal_length', data=dataset)
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.stripplot(x='species', y='sepal_length', data=dataset, color='blue', jitter=False)
plt.title("Strip Plot of Sepal Length by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
sns.swarmplot(x='species', y='sepal_length', data=dataset, color='blue')
plt.title("Swarm Plot of Sepal Length by Species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
dataset_numeric = dataset.drop(columns='species')
correlation_matrix = dataset_numeric.corr()
sns.clustermap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('iris')
dataset_numeric = dataset.drop(columns='species')
correlation_matrix = dataset_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heat Map of Feature Correlations")
plt.show()
```
