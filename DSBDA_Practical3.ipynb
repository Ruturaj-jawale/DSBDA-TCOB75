```python
import pandas as pd
import numpy as np
df = pd.read_csv("iris.csv")
df
```

```python
print("mean sepal length:",df["sepal_length"].mean())
print("mean sepal width:",df["sepal_width"].mean())
print("mean petal length:",df["petal_length"].mean())
print("mean petal width:",df["petal_width"].mean())
```

```python
print("median sepal length:",df["sepal_length"].median())
print("median sepal width:",df["sepal_width"].median())
print("median petal length:",df["petal_length"].median())
print("median petal width:",df["petal_width"].median())
```

```python
print("mode sepal length:",df["sepal_length"].mode())
print("mode sepal width:",df["sepal_width"].mode())
print("mode petal length:",df["petal_length"].mode())
print("mode petal width:",df["petal_width"].mode())
```

```python
print("standart deviation of sepal length:",df["sepal_length"].std())
print("standart deviation of sepal width:",df["sepal_width"].std())
print("standart deviation of petal length:",df["petal_length"].std())
print("standart deviation of petal width:",df["petal_width"].std())
```

```python
print("minimum sepal length:",df["sepal_length"].min())
print("minimum sepal width:",df["sepal_width"].min())
print("minimum petal length:",df["petal_length"].min())
print("minimum petal width:",df["petal_width"].min())
```

```python
print("maximum sepal length:",df["sepal_length"].max())
print("maximum sepal width:",df["sepal_width"].max())
print("maximum petal length:",df["petal_length"].max())
print("maximum petal width:",df["petal_width"].max())
```

```python
array1 = np.array(df["sepal_length"])
array2 = np.array(df["sepal_width"])
array3 = np.array(df["petal_length"])
array4 = np.array(df["petal_width"])
print("maximum sepal length:",max(array1))
print("maximum sepal width:",max(array2))
print("maximum petal length:",max(array3))
print("maximum petal width:",max(array4))
print("minimum sepal length:",min(array1))
print("minimum sepal width:",min(array2))
print("minimum petal length:",min(array3))
print("minimum petal width:",min(array4))
```
